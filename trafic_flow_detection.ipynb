```python
# import required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.ensemble import IsolationForest
```

```python
df=pd.read_csv('system_network_security_dataset.csv')
df.head()
```

```python

df.isnull().sum()
```

```python
df.info()
```

```python
# Check missing value
df.isnull().sum()
```

```python
## step 2: Data Preprocessing
features=df.drop('label',axis=1)
```

```python
for col in features.columns:
    if df[col].dtype == 'bool':
        df[col]=df[col].astype(int)
```

```python
features.info()
```

```python
## Feature Engineering 

from sklearn.preprocessing import StandardScaler

s=StandardScaler()
sf=s.fit_transform(features)
```

```python
scaled_feature_df=pd.DataFrame(sf,columns=features.columns)
scaled_feature_df.head()
```

```python
features.columns
```

```python
# parameters for isolation forest
n_estimators=100
contamination = 0.04
sample_size=256
```

```python
# train the isolation forest model

model=IsolationForest(n_estimators=n_estimators,contamination=contamination,max_samples=sample_size,random_state=42)
model.fit(scaled_feature_df)
anomaly_labels=model.predict(scaled_feature_df) # -1 for anomalies and 1 for normal instances
scaled_feature_df['anomaly']=anomaly_labels
scaled_feature_df.head()
scaled_feature_df['anomaly'].value_counts()
plt.figure(figsize=(10,8))
sns.scatterplot(data=scaled_feature_df,x='packet_size',y='packet_count_5s',hue='anomaly')
plt.show()
```
